```python
# Import SQLAlchemy `automap` and other related dependencies here
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
import sqlite3
```

```python
# Import Matplot lib
import matplotlib
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt

#Import Pandas
import pandas as pd
```

```python
#Creating an engine for the database
#need to figure out how to remove the password
engine = create_engine('postgresql://postgres:postgres@localhost:5432/PewlettHackardHR')

#creating a connection for the database
connection = engine.connect()
```

```python
#Using Inspector to make sure the tables were imported
inspector=inspect(engine)
inspector.get_table_names()
```

```python
# Importing the employees table into Pandas
ee_data = pd.read_sql("SELECT * FROM employees", connection)

ee_data.head(20)
```

```python
# Importing the salaries table into Pandas
salary_df = pd.read_sql("SELECT * FROM salaries", connection)

salary_df.head(20)
```

```python
# Importing the titles table into Pandas
titles_df = pd.read_sql("SELECT * FROM titles", connection)

titles_df.head(20)
```

What are the most common salary ranges for employees?

```python
#merging ee_data and salaries 
ee_salaries=pd.merge(ee_data, salary_df, on= "emp_no")

ee_salaries.head(20)
```

```python
#renmaming the emp_title_id column
ee_salaries2=ee_salaries.rename(columns={"emp_title_id" : "title_id"})

ee_salaries2
```

```python
#finding null data
ee_salaries2.isnull().values.any()
```

```python
# aggregating summary statistics for salaries
agg_salaries = round(pd.DataFrame(ee_salaries2.agg({'salary': ['mean', 'median', 'std']})),2)

agg_salaries
```

Histogram displaying the distribution of salary ranges for employees

```python
#histogram the the most common salary ranges for employees
plt.hist(ee_salaries2['salary'])
plt.xlabel('Average Salary')
plt.ylabel('Number of Employees')
plt.show()
```

What is the average salary by title?

```python
#merging ee_data and titles
ee_sal_titles=pd.merge(ee_salaries2, titles_df, on= "title_id")

ee_sal_titles
```

Generating a bar char tshowing the average salary by job title

```python
#filter dataframe to only timepoint and drug regimen
grouped_title = ee_sal_titles.groupby(["title"])

#number of timepoints
avg_salary = grouped_title["salary"].mean()

#chart the data
salary_chart = avg_salary.plot(kind="bar", title="Average Salary by Job Title", color = "blue", alpha=0.75)
salary_chart.set_xlabel("Job Title")
salary_chart.set_ylabel("Average Salary")

plt.tight_layout()
plt.show();
```
